```python
# Store filepath in a variable
#Creating dataframes for every month
January17 = pd.read_csv('201701-citibike-tripdata.csv')
February17 = pd.read_csv('201702-citibike-tripdata.csv')
March17 = pd.read_csv('201703-citibike-tripdata.csv')
April17 = pd.read_csv('201704-citibike-tripdata.csv')
May17 = pd.read_csv('201705-citibike-tripdata.csv')
June17 = pd.read_csv('201706-citibike-tripdata.csv')
July17 = pd.read_csv('201707-citibike-tripdata.csv')
August17 = pd.read_csv('201708-citibike-tripdata.csv')
September17 = pd.read_csv('201709-citibike-tripdata.csv')
October17 = pd.read_csv('201710-citibike-tripdata.csv')
November17 = pd.read_csv('201711-citibike-tripdata.csv')
December17 = pd.read_csv('201712-citibike-tripdata.csv')
```
